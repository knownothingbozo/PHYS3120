```python
def s(x, y, n):
    p = x[-1] - x[0]
    dx = x[1] - x[0]
    rs = 0
    for i in y:
        rs += i
    a0 = (1/p)*rs*dx
    a = []
    b = []
    for j in range(1, n+1):
        an = 0
        bn = 0
        m = 0
        for k in y:
            an += (2/p)*k*np.cos((2*np.pi*j*x[m])/p)*dx
            bn += (2/p)*k*np.sin((2*np.pi*j*x[m])/p)*dx
            m += 1
        a.append(an)
        b.append(bn)
    l = 0
    z = []
    for i in y:
        c = 0
        w = 0
        for v in range(1, n+1):
            c += (a[w] * np.cos((2*np.pi*v*x[l])/p) + b[w] * np.sin((2*np.pi*v*x[l])/p))
            w += 1
        l += 1
        z.append(a0 + c)
        
    return z
```

```python
def fsin(x, y, n):
    p = x[-1] - x[0]
    dx = x[1] - x[0]
    b = []
    for j in range(1, n+1):
        bn = 0
        m = 0
        for k in y:
            bn += (2/p)*k*np.sin((2*np.pi*j*x[m])/p)*dx
            m += 1
        b.append(bn)
    l = 0
    z = []
    for i in y:
        c = 0
        w = 0
        for v in range(1, n+1):
            c += (b[w] * np.sin((2*np.pi*v*x[l])/p))
            w += 1
        l += 1
        z.append(c)
        
    return z
```

```python
def fcos(x, y, n):
    p = x[-1] - x[0]
    dx = x[1] - x[0]
    rs = 0
    for i in y:
        rs += i
    a0 = (1/p)*rs*dx
    a = []
    for j in range(1, n+1):
        an = 0
        bn = 0
        m = 0
        for k in y:
            an += (2/p)*k*np.cos((2*np.pi*j*x[m])/p)*dx
            m += 1
        a.append(an)
    l = 0
    z = []
    for i in y:
        c = 0
        w = 0
        for v in range(1, n+1):
            c += (a[w] * np.cos((2*np.pi*v*x[l])/p))
            w += 1
        l += 1
        z.append(a0 + c)
        
    return z
```

```python
t = np.linspace(0, 1, num=1000, endpoint=False)
x = []
n = 0.5
for i in t:
    if i < n:
        x.append(1)
    elif i < n+3:
        x.append(0)
```

```python
i = 1
y = s(t, x, i)
y1 = fsin(t,x,i)
y2 = fcos(t,x,i)

fig = plt.figure(figsize=(15,13), dpi=75)

ax = fig.add_subplot(2, 2, 1)
ax.annotate("Iterations: {0}".format(i), xy=(0,1.02), xycoords='axes fraction', fontsize=15)
ax.plot(t, x)

ax = fig.add_subplot(2, 2, 2)
ax.annotate('Fourier Series', xy=(0,1.02), xycoords='axes fraction', fontsize=15)
ax.plot(t, y)

ax = fig.add_subplot(2, 2, 3)
ax.annotate('Fourier Sine Series', xy=(0,1.02), xycoords='axes fraction', fontsize=15)
ax.plot(t, y1)

ax = fig.add_subplot(2, 2, 4)
ax.annotate('Fourier Cosine Series', xy=(0,1.02), xycoords='axes fraction', fontsize=15)
ax.plot(t, y2)
```

```python
def Ft(t, s, h):
    gel = []
    N = len(t)
    alpha = t
    count = range(N)
    for i in count:
        bar = 0
        for j in count:
            foo = s[j]*np.cos((t[j]*alpha[i]))
            bar += foo
        gel.append(abs(bar))       
    return (t, gel)
```

```python
step = 0.01
t = np.arange(0,50,step)
s = np.sin(5*t) + np.sin(15*t) + np.sin(20*t)

x, y = Ft(t, s, step)
```

```python
fig, axs = plt.subplots()
axs.clear()

## Variables that define the wave
start = 0
end = 100
periods = 50
steps_per_period = 99   ##This can't go below 4
                        ## If this value is not divisible by four the waveform will not end in one full cycle
stepnumber = (steps_per_period * periods) + 1 # for neat waves
amplitude = 4
direction = 0 # starts the wave up or down. I could have made this into a phase shift parameter.
                # I don't see the point for this application.
                # you could either fiddle with the incrementation code or cheat by chopping and shifting
##

def triangle(start, end, stepnumber, period, amplitude, direction):
    x = np.linspace(start, end, stepnumber)
    y = []
#     print(x)
    Nst = (stepnumber - 1)
#     step = (end - start)/ Nst
    
    v = (Nst / period)
    while v % 4 != 0:
        v+=1
    inc = amplitude / (v/4)
        
#     print(v)    ## Uncomment for debug
#     print(Nst)
    
    if direction == 0:
        j = 0
        l = 0
        for i in x:
#             print(l)    ## Uncomment for debug
            if l == 0:
                y.append(j)
                l += 1
            elif l <= v/4:
                j += inc
                y.append(j)
                l += 1
            elif l == (v-1) and l == v/2 + v/4:
                j -= inc
                y.append(j)
                l=0
                j=0
            elif l > v/4 and l <= v/2 + v/4:
                j -= inc
                y.append(j)
                l += 1
            elif l > v/2 + v/4 and l < (v-1):
                j += inc
                y.append(j)
                l += 1
            elif l == (v-1):
                j += inc
                y.append(j)
                l=0
                j=0
            
            
    elif direction == 1:
        j = 0
        l = 0
        m = 0
        for i in x:
#             print(l)    ## Uncomment for debug
            if l == 0:
                y.append(j)
                l += 1
            elif l <= v/4:
                j -= inc
                y.append(j)
                l += 1
            elif l == (v-1) and l == v/2 + v/4:
                j += inc
                y.append(j)
                l=0
                j=0
            elif l > v/4 and l <= v/2 + v/4:
                j += inc
                y.append(j)
                l += 1
            elif l > v/2 + v/4 and l < (v-1):
                j -= inc
                y.append(j)
                l += 1
            elif l == (v-1):
                j -= inc
                y.append(j)
                l=0
                j=0
                
    return x, y
        
x, y = triangle(start, end, stepnumber, periods, amplitude, direction)
axs.plot(x, y)
```
